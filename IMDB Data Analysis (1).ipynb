```python
#title of the movie having runtime>= 180 min.
data[data['Runtime (Minutes)']>=180]['Title']
```

```python
#highest average votting
data.groupby('Year')['Votes'].mean().idxmax()
```

```python
#highest average votting
data.groupby('Year')['Votes'].mean().sort_values(ascending=False)
```

```python
sns.barplot(x='Year',y='Votes',data=data)
plt.title('Votes By Year')
plt.show
```

```python
#HIGHEST AVERAGE REVANUE IN WHICH YEAR
data.groupby('Year')['Revenue (Millions)'].mean().sort_values(ascending=False)
```

```python
sns.barplot(x='Year',y='Revenue (Millions)',data=data)
plt.title('revanue By Year')
plt.show
```

```python
data.columns
```

```python
#Top 10 lengthy movie title and Runtime
top_10=data.nlargest(10,'Runtime (Minutes)')[['Title','Runtime (Minutes)']]\
.set_index('Title')
top_10
```

```python
sns.barplot(x='Runtime (Minutes)',y=top_10.index,data=top_10)
plt.title('Top 10 lengthy movie title and Runtime')
plt.show()
```

```python
#Highest Rated Movies with Directors & Actor
top_10R=data.nlargest(10,'Rating')[['Title','Director','Actors','Rating']]\
.set_index('Title')
top_10R
```

```python
sns.barplot(x='Rating',y=top_10R.index,data=top_10R,hue='Director',dodge=False)
plt.legend(bbox_to_anchor=(1.05,1),loc=2)
plt.show()
```
